### Task 2. Drop Useless Columns
Useless columns - ['dob','sofifa_id', 'player_url', 'long_name', 'body_type',
'real_face', 'loaned_from', 'nation_position', 'nation_jersey_number']

```python
#Droping unneccessary columns which will be of no use for analysis

df1= df.drop(['dob','sofifa_id', 'player_url', 'long_name', 'body_type',
              'real_face', 'loaned_from', 'nation_position', 'nation_jersey_number'], axis= 1)
```

### Task 3. Data Preprocessing - Part 1
- Calculate Body Mass Index (BMI) of each player
- Display first 10 rows of Player Name with its Position

##### As we know the formula of BMI is as follow:
      Formula: weight (kg) / [height (m)]2
      so we will use this to calculate BMI

```python
#Calculating BMI of Players

df1['BMI']= df1['weight_kg']/ df['height_cm']**2
```

```python
#Here BMI is in Kg/cm^2

df1.BMI
```

```python
#We are converting BMI in Kg/m^2

df1['BMI']= df1['BMI']*10000
```

```python
df1[["short_name", "BMI"]]
```

```python
#Players with their respective positions

df2= df1[['short_name', 'player_positions']]
df2
```

### Task 4. Data Preprocessing - Part 2
- Create DataFrame of each field positions and add prefix as (Position_)
- Concate above DataFrame with our original DataFrame

```python
#Creating Dataframe for different positions on field

data= {"position_goalkeeper" : 'GK',
        "position_defenders" : ['Center Back(CB)', 'Fullback (LB, RB)', 'Wingback (LWB, RWB)', 'Sweeper (SW)'],
        "position_midfield" : ['Defensive Midfielder (DM)', 'Central Midfielder (CM)', 'Attacking Midfielder (AM)', 'Left/Right Midfielder (LM, RM)'],
        "position_offensive" : ['Center Forward (CF)', 'Striker (S)', 'Second Striker (SS)']}

df3= pd.DataFrame({key:pd.Series(value) for key, value in data.items()})

df3
```

```python
#concating above dataframe with original dataframe

df4= pd.concat([df1,df3], axis=1)
df4
```

### Task 5. Data Preprocessing - Part 3
- Clean Position Ratings Column
    - Rating Columns - ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram','lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb','lcb', 'cb', 'rcb', 'rb']
    - Fill Empty values = 0
    - Make Data-type as = Integer
    
    
- Fill Missing Values of Skills
    - Skills - ["dribbling", "defending", "physic", "passing", "shooting", "pace"]
    - Fill Missing Values = As Median

```python
#Filling empty values of these columns with Zero

df4[["ls","st", "rs", "lw","lf"]]= df4[["ls","st", "rs", "lw","lf"]].fillna(value= 0)
df4[["cf", "rf", "rw", "lam", "cam"]]= df4[["cf", "rf", "rw","lam", "ram"]].fillna(value= 0)
df4[["ram", "lm", "lcm", "cm", "rcm"]]= df4[["ram", "lm", "lcm", "cm", "rcm"]].fillna(value= 0)
df4[["rm", "lwb", "ldm", "cdm", "rdm"]]= df4[["rm", "lwb", "ldm", "cdm", "rdm"]].fillna(value= 0)
df4[["rwb", "lb", "lcb", "cb", "rcb", "rb"]]= df4[["rwb", "lb", "lcb", "cb", "rcb", "rb"]].fillna(value= 0)
```

```python
df4.tail(10)
```

```python
#Filling missing values of skills with their respective median

dribbling_med= df4["dribbling"].median()
defending_med= df4["defending"].median()
physic_med= df4["physic"].median()
passing_med= df4["passing"].median()
shooting_med= df4["shooting"].median()
pace_med= df4["passing"].median()

print("Median of different skills respectively")
print(dribbling_med)
print(defending_med)
print(physic_med)
print(passing_med)
print(shooting_med)
print(pace_med)

df4[["dribbling"]]= df4[["dribbling"]].fillna(value= dribbling_med)
df4[["defending"]]= df4[["defending"]].fillna(value= defending_med)
df4[["physic"]]= df4[["physic"]].fillna(value= physic_med)
df4[["passing"]]= df4[["passing"]].fillna(value= passing_med)
df4[["shooting"]]= df4[["shooting"]].fillna(value= shooting_med)
df4[["pace"]]= df4[["pace"]].fillna(value= pace_med)
```
